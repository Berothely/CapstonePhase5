- des valeurs incoh√©rentes comme ‚Äú3+‚Äù et ‚ÄúALL‚Äù.

Nous commen√ßons donc par corriger ces √©l√©ments et d'agir sur d'autres elements.


```python
# Suppression de la premi√®re ligne descriptive
ipc_data = ipc_data.iloc[1:]

# Suppression des valeurs non pertinentes
ipc_data = ipc_data[~ipc_data['Phase'].isin(['ALL'])]

ipc_data['Phase'] = ipc_data['Phase'].replace('3+', 4)

# Suppression des colonnes non pertinentes
cols_drop_ipc = ['Country','Validity period', 'From', 'To', 'iso3', 'Area', 'Percentage']
ipc_data = ipc_data.drop(columns=[c for c in cols_drop_ipc if c in ipc_data.columns])

# Renommage des colonnes
ipc_data = ipc_data.rename(columns={
    'Date of analysis': 'Date',
    'Total country population': 'Population totale',
    'Level 1': 'Departement',
    'Number' : 'Population touchee'
})


# Conversion du type de la colonne Phase
ipc_data['Phase'] = pd.to_numeric(ipc_data['Phase'], errors='coerce')

#correction de la region de ZMPAP figurant comme un departement

ipc_data["Departement"] = (
    ipc_data["Departement"]
    .str.strip()
    .replace({
        "ZMPAP":"Ouest"
    })
)

ipc_data.dropna(inplace=True)

ipc_data
```

# Phase 4 ‚Äî Mod√©lisation (Modeling)

Cette phase vise √† construire et √† √©valuer un **mod√®le pr√©dictif supervis√©**
permettant d‚Äôestimer la **phase d‚Äôins√©curit√© alimentaire (IPC)** √† partir des
indicateurs issus du *Joint Monitoring Report (JMR)*.

La d√©marche suit une approche scientifique en quatre √©tapes :
1. S√©paration des donn√©es en features et target ;  
2. Division du jeu de donn√©es en ensembles d‚Äôentra√Ænement et de test ;

## üßπ 4.1 ‚Äì Harmonisation et agr√©gation des doublons

Avant toute mod√©lisation, il a √©t√© observ√© que certaines communes apparaissaient plusieurs fois
pour la m√™me **date**, **phase**, **d√©partement** et **indicateurs**, ne diff√©rant que par la variable
**Population touch√©e**.

Pour garantir une coh√©rence territoriale et statistique :

- Les **indicateurs num√©riques** (GLM, NDVI, pluie, taux de change, prix) ont √©t√© agr√©g√©s par **moyenne** ;
- La **population touch√©e** a √©t√© agr√©g√©e par **somme**, repr√©sentant l‚Äôensemble des personnes affect√©es dans la commune ;
- La **phase IPC** (1 √† 5) a √©t√© conserv√©e telle quelle.

Ainsi, chaque **commune-date-phase** devient une observation unique, propre √† la mod√©lisation.


### D√©finition des variables cibles et explicatives

Avant de passer √† la mod√©lisation, il est essentiel d‚Äôidentifier les colonnes
qui serviront de variables **explicatives (features)** et **cibles (target)**.


-- Variable cible (target) :
- **`Phase`** ‚Üí Niveau d‚Äôins√©curit√© alimentaire (1 √† 5) selon le syst√®me IPC.

-- Variables explicatives (features) :

| Indicateur | Description | Dimension |
|-------------|--------------|------------|
| **GLM** | Indice global combin√© climatique | Environnement |
| **Drought rainfall** | Quantit√© ou indice de pr√©cipitation | Climat |
| **Drought NDVI** | Indice de v√©g√©tation (sant√© des cultures) | Climat |
| **Exchange rates** | Taux de change Gourde/USD | √âconomie |
| **Food Prices** | Prix moyen des denr√©es alimentaires | √âconomie |
| **Target** | Indicateur de seuil ou population cibl√©e | Social |

Ces indicateurs forment la base du raisonnement explicatif du mod√®le.


###  Explication :
Chaque ligne du dataset repr√©sente d√©sormais **un d√©partement √† une date donn√©e**,
avec la valeur moyenne de chaque indicateur.  

Les colonnes obtenues (NDVI, pluie, prix, taux de change, etc.) 
servent √† expliquer pourquoi la **phase IPC (1 √† 5)** atteint tel niveau.

Ainsi, le mod√®le cherche √† **√©tablir un lien causal entre les indicateurs et la phase IPC** :
> ‚ÄúQuels facteurs expliquent la situation actuelle d‚Äôun d√©partement ?‚Äù


```python
# Pivot du dataset : chaque indicateur devient une colonne
data_pivot = (
    final_data
    .pivot_table(
        index=['Code_Postal', 'Departement', 'Commune', 'Date_x', 'Phase', 'Population touchee'],
        columns='Indicateur',
        values='Valeur',
        aggfunc='mean'
    )
    .reset_index()
)

# V√©rification du r√©sultat
print("Dimensions apr√®s pivot :", data_pivot.shape)
display(data_pivot.head())
```

```python
# Agr√©gation par commune, d√©partement, date et phase
data_pivot['Annee'] = pd.to_datetime(data_pivot['Date_x']).dt.year

data_clean = (
    data_pivot
    .groupby(['Departement', 'Commune', 'Annee'], as_index=False)
    .agg({
        'Phase': 'mean',
        'GLM': 'mean',
        'Drought NDVI': 'mean',
        'Drought rainfall': 'mean',
        'Exchange rates': 'mean',
        'Food prices': 'mean',
        'Target': 'mean'
    })
)

print("Dimensions apr√®s agr√©gation :", data_clean.shape)
data_clean
```

```python
# Suppression des enregistrements sans phase IPC
final_data = data_clean.dropna(subset=['Target','GLM', 'Drought NDVI', 'Drought rainfall', 'Exchange rates', 'Food prices'])
final_data
```

```python
# === Supervised REGRESSION on 'Phase' ‚Äî Multiple Linear Regression + RandomForest ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

RANDOM_STATE = 42

# 1) Target & features (numeric-only)
df = final_data.copy()
y = pd.to_numeric(df["Phase"], errors="coerce")

# drop obvious non-features if present
X = df.drop(columns=["Phase", "Departement", "Commune", "Annee"], errors="ignore")

# keep only numeric columns (handles GLM, Drought NDVI, Drought rainfall, Exchange rates, Food prices, Target if you keep it)
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
X = X[num_cols]

# 2) Preprocess: median impute + scale (good for LinearRegression; RF is robust but harmless to scale)
preprocess = ColumnTransformer(
    [("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc", StandardScaler())]), num_cols)],
    remainder="drop"
)

# 3) Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

# 4) Models
models = {
    "MultipleLinearRegression": Pipeline([("prep", preprocess), ("mdl", LinearRegression())]),
    "RandomForestRegressor":    Pipeline([("prep", preprocess),
                                          ("mdl", RandomForestRegressor(
                                              n_estimators=500, random_state=RANDOM_STATE, n_jobs=-1
                                          ))]),
}

# 5) Fit ‚Üí Predict ‚Üí Evaluate
results = []
for name, pipe in models.items():
    print(f"\n=== {name} ===")
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    mae  = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2   = r2_score(y_test, y_pred)
    print(f"MAE:  {mae:.4f} | RMSE: {rmse:.4f} | R¬≤: {r2:.4f}")

    # Quick residuals check
    resid = y_test - y_pred
    plt.figure(figsize=(5.2,4))
    plt.scatter(y_pred, resid, alpha=0.6)
    plt.axhline(0, ls="--", linewidth=1)
    plt.xlabel("Predicted"); plt.ylabel("Residuals")
    plt.title(f"Residuals ‚Äî {name}")
    plt.tight_layout()
    plt.show()

    results.append({"model": name, "MAE": mae, "RMSE": rmse, "R2": r2})

# 6) Summary (lower RMSE is better)
summary = pd.DataFrame(results).set_index("model").sort_values("RMSE")
print("\n=== Summary (lower RMSE is better) ===")
display(summary)
```

```python
# Choix du meilleur mod√®le bas√© sur les performances
best_model = models["RandomForestRegressor"]

# Confirmation
print("Le mod√®le final s√©lectionn√© est : RandomForestRegressor")
```

```python
# === Cellule : split entrainement/test ===
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train distribution:\n", y_train.value_counts(normalize=True))
```

```python
# Pick winner, quick CV check, feature importance, save
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.model_selection import cross_val_score

# 1) Pick the fitted RF pipeline from your previous cell
rf_pipe = models["RandomForestRegressor"]   # already fitted

# Recompute test predictions (optional)
y_pred = rf_pipe.predict(X_test)

# 2) Quick cross-validation RMSE on the TRAIN set (refits internally)
cv_rmse = -cross_val_score(
    rf_pipe, X_train, y_train,
    scoring="neg_root_mean_squared_error",
    cv=5, n_jobs=-1
)
print(f"CV RMSE (5-fold): mean={cv_rmse.mean():.4f} | std={cv_rmse.std():.4f}")

# 3) Feature importance (top n)
try:
    feat_names = rf_pipe.named_steps['prep'].get_feature_names_out()
except AttributeError:
    feat_names = ['GLM', 'Drought NDVI', 'Drought rainfall', 'Exchange rates', 'Food prices', 'Target']
importances = rf_pipe.named_steps["mdl"].feature_importances_
imp_df = (pd.DataFrame({"feature": feat_names, "importance": importances})
            .sort_values("importance", ascending=False)
            .head(10))
```

# Phase 5 ‚Äî √âvaluation et validation du mod√®le

Dans cette phase, on a v√©rifier si le mod√®le qu'on a construit est r√©ellement fiable pour anticiper le niveau d‚Äôins√©curit√© alimentaire (Phase IPC) dans les d√©partements et communes d‚ÄôHa√Øti.

L‚Äôobjectif n‚Äôest pas seulement d‚Äôavoir un bon score math√©matique. L‚Äôobjectif est de pouvoir r√©pondre √† une question op√©rationnelle tr√®s simple :

> Est-ce que nous pouvons utiliser ce mod√®le pour dire √† une autorit√© (Agriculture, CNSA, PAM) :

La phase 5 n‚Äôest donc pas seulement technique. C‚Äôest la phase o√π on juge si le mod√®le peut vivre dans le monde r√©el.


## 5.1 Performance pr√©dictive du mod√®le sur des donn√©es jamais vues

Dans cette section, on teste le mod√®le sur des donn√©es qu‚Äôil n‚Äôa pas vues pendant l‚Äôentra√Ænement.  

On mesure trois choses :
- **MAE (Mean Absolute Error)** : l‚Äôerreur moyenne absolue entre la phase r√©elle et la phase pr√©dite.
- **RMSE (Root Mean Squared Error)** : p√©nalise plus fort les grosses erreurs.
- **R¬≤** : quelle part de la variation de la phase IPC est expliqu√©e par nos indicateurs.

Plus MAE et RMSE sont bas, mieux c‚Äôest.  
Plus R¬≤ est haut, mieux c‚Äôest.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Pr√©dictions du mod√®le sur le jeu de test
y_pred = best_model.predict(X_test)

# Calcul des m√©triques de performance
mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

print("=== √âvaluation du mod√®le sur le jeu de test ===")
print(f"MAE :  {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R¬≤   : {r2:.4f}")
```

- Ce faible MAE veut dire que, en moyenne, le mod√®le ne se trompe pas beaucoup sur la phase IPC r√©elle de la commune.  
- Le R¬≤  de 0.82 signifie que nos indicateurs (prix alimentaires, taux de change, pluie, s√©cheresse NDVI...) expliquent d√©j√† plus de 3/4 de la gravit√© alimentaire observ√©e sur le terrain.  
  C‚Äôest un bon r√©sultat solide pour un pays en crise multidimensionnelle.
  
En clair : le mod√®le est capable de donner un signal cr√©dible d‚Äôalerte.


## 5.2 Visualisation directe : r√©alit√© vs pr√©diction

Un bon mod√®le ne doit pas √™tre √©valu√© uniquement par des chiffres.  
Ici, on veux voir visuellement si les pr√©dictions suivent bien la r√©alit√©.

On a fait deux v√©rifications :

1. **R√©alit√© vs Pr√©diction**
   - Chaque point du graphe repr√©sente une commune sur une ann√©e donn√©e.
   - Vu que les points sont proches de la diagonale rouge (ligne parfaite), √ßa veut dire que le mod√®le colle √† la r√©alit√©.

2. **Analyse de l‚Äôerreur (r√©sidus)**
   - On regarde l‚Äôerreur `IPC_r√©el - IPC_pr√©dit`.

## 5.3 V√©rification du surapprentissage (overfitting)

On doit v√©rifier si le mod√®le est honn√™te.

Un mod√®le peut ‚Äútricher‚Äù : il peut apprendre par c≈ìur les donn√©es historiques (train), avoir l‚Äôair parfait, mais devenir inutile d√®s qu‚Äôon lui donne une nouvelle p√©riode ou une nouvelle commune.

Pour v√©rifier √ßa :
- on mesure l‚Äôerreur sur l‚Äô√©chantillon d‚Äôentra√Ænement (donn√©es connues),
- on mesure l‚Äôerreur sur l‚Äô√©chantillon de test (donn√©es jamais vues),
- on compare.

Si l‚Äôerreur explose sur le test, c‚Äôest du surapprentissage.


```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Pr√©dictions train et test
y_train_pred = best_model.predict(X_train)
y_test_pred  = best_model.predict(X_test)

# RMSE train / test
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test  = np.sqrt(mean_squared_error(y_test,  y_test_pred))

print("RMSE (train) :", round(rmse_train, 4))
print("RMSE (test)  :", round(rmse_test, 4))

if rmse_test - rmse_train > 0.05:
    print("Le mod√®le est probablement en surapprentissage (il apprend trop le pass√©).")
else:
    print("Pas de signe majeur de surapprentissage : le mod√®le reste fiable sur de nouvelles communes/p√©riodes.")
```

## 5.4 Validation crois√©e 

Maintenant on veut r√©pondre √† une question de confiance :  
Est-ce que le mod√®le reste bon si on change l√©g√®rement les donn√©es d‚Äôentra√Ænement ?  
Ou bien est-ce qu‚Äôil est fragile, c‚Äôest-√†-dire performant uniquement dans certains d√©partements mais pas dans d‚Äôautres ?

Pour √ßa, on utilise la validation crois√©e (5-fold cross-validation) : on r√©-entraine et r√©√©value le mod√®le plusieurs fois sur des sous-√©chantillons diff√©rents, puis on regarde la stabilit√© des scores R¬≤.


```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='r2')

print("Scores R¬≤ par fold :", cv_scores)
print(f"R¬≤ moyen     : {cv_scores.mean():.4f}")
print(f"√âcart-type   : {cv_scores.std():.4f}")
```
